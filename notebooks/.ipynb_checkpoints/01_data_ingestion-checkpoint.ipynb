```python
import pandas as pd

# Load fund master
df = pd.read_csv("../data/raw/01_fund_master.csv")

print("Unique Fund Houses:")
print(df['fund_house'].unique())

print("\nCategories:")
print(df['category'].value_counts())

print("\nSub-categories:")
print(df['sub_category'].value_counts())

print("\nRisk Grades:")
print(df['risk_category'].value_counts())
```

```python
# Validate AMFI codes
fund_master = pd.read_csv("../data/raw/01_fund_master.csv")
nav_history = pd.read_csv("../data/raw/02_nav_history.csv")

master_codes = set(fund_master['amfi_code'])
nav_codes = set(nav_history['amfi_code'])

missing = master_codes - nav_codes

if missing:
    print(f"⚠️ {len(missing)} codes missing in nav_history:")
    print(missing)
else:
    print("✅ All AMFI codes validated! No mismatches found.")

print(f"\nTotal funds in master: {len(master_codes)}")
print(f"Total funds in nav_history: {len(nav_codes)}")
```
