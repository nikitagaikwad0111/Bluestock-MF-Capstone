```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# Load data
df_nav = pd.read_csv("../data/processed/clean_nav.csv")
df_nav['date'] = pd.to_datetime(df_nav['date'])

df_fund = pd.read_csv("../data/raw/01_fund_master.csv")

print(f"Nav Shape: {df_nav.shape}")
print(f"Fund Master shape: {df_fund.shape}")
```

```python
# Sort and compute daily returns
df_nav = df_nav.sort_values(['amfi_code', 'date'])
df_nav['daily_return'] = df_nav.groupby('amfi_code')['nav'].pct_change()

# Drop NaN returns
df_returns = df_nav.dropna(subset=['daily_return'])
print(df_returns.head())
```

```python
# Compute VaR (95%) and CVaR for each fund
var_cvar_results = []

for code, group in df_returns.groupby("amfi_code"):
    returns = group["daily_return"]

    # VaR = 5th percentile
    var_95 = np.percentile(returns, 5)

    # CVaR = mean of returns below VaR
    cvar_95 = returns[returns <= var_95].mean()

    var_cvar_results.append({
        "amfi_code": code,
        "var_95_pct": round(var_95 * 100, 4),
        "cvar_95_pct": round(cvar_95 * 100, 4)
    })

var_cvar_df = pd.DataFrame(var_cvar_results)

# Merge with fund names
var_cvar_df = var_cvar_df.merge(
    df_fund[["amfi_code", "scheme_name", "category"]],
    on="amfi_code"
)

print(var_cvar_df.head(10))
```

```python
# Save report
output_path = "../data/processed/var_cvar_report.csv"
var_cvar_df.to_csv(output_path, index=False)
print(f"✅ Saved: {output_path}")
print(var_cvar_df.sort_values("var_95_pct").head(10))
```

```python
# Select 5 key funds
five_funds = {
    119551: "SBI Bluechip",
    119598: "SBI Small Cap",
    125497: "HDFC Top 100",
    120503: "ICICI Bluechip",
    118632: "Nippon Large Cap"
}

# Filter returns for these 5 funds
df_5funds = df_returns[df_returns["amfi_code"].isin(five_funds.keys())].copy()

# Compute rolling 90-day Sharpe (Rf = 6.5% annual = 0.065/252 daily)
Rf_daily = 0.065 / 252

rolling_sharpe_list = []

for code, name in five_funds.items():
    fund_data = df_5funds[df_5funds["amfi_code"] == code].copy()
    fund_data = fund_data.sort_values("date")
    
    roll_mean = fund_data["daily_return"].rolling(90).mean()
    roll_std = fund_data["daily_return"].rolling(90).std()
    
    fund_data["rolling_sharpe"] = (roll_mean - Rf_daily) / roll_std * np.sqrt(252)
    fund_data["fund_name"] = name
    rolling_sharpe_list.append(fund_data)

rolling_sharpe_df = pd.concat(rolling_sharpe_list)
print(rolling_sharpe_df.head())
```

```python
# Plot Rolling 90-day Sharpe for 5 funds
plt.figure(figsize=(14, 6))

for code, name in five_funds.items():
    fund_data = rolling_sharpe_df[rolling_sharpe_df["amfi_code"] == code]
    plt.plot(fund_data["date"], fund_data["rolling_sharpe"], label=name)

plt.axhline(y=0, color="black", linestyle="--", linewidth=0.8, label="Sharpe = 0")
plt.axhline(y=1, color="green", linestyle="--", linewidth=0.8, label="Sharpe = 1 (Good)")

plt.title("Rolling 90-Day Sharpe Ratio — 5 Key Funds (2022–2025)")
plt.xlabel("Date")
plt.ylabel("Sharpe Ratio")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()

# Save chart
chart_path = "../reports/rolling_sharpe_chart.png"
plt.savefig(chart_path, dpi=150)
plt.show()
print(f"✅ Saved: {chart_path}")
```

```python
# Load transactions
df_tx = pd.read_csv("../data/processed/clean_transactions.csv", 
                     parse_dates=["transaction_date"])

# Find first transaction year per investor
first_tx = df_tx.groupby("investor_id")["transaction_date"].min().reset_index()
first_tx["cohort_year"] = first_tx["transaction_date"].dt.year
first_tx.columns = ["investor_id", "first_transaction_date", "cohort_year"]

# Merge back with transactions
df_tx = df_tx.merge(first_tx[["investor_id", "cohort_year"]], on="investor_id")

# Filter SIP only for SIP amount analysis
sip_only = df_tx[df_tx["transaction_type"] == "SIP"]

# Cohort analysis
cohort = df_tx.groupby("cohort_year").agg(
    num_investors=("investor_id", "nunique"),
    avg_sip_amount=("amount_inr", "mean"),
    total_invested=("amount_inr", "sum"),
    top_fund=("amfi_code", lambda x: x.mode()[0])
).reset_index()

print(cohort)
```

```python
# Filter investors with SIP transactions only
sip_tx = df_tx[df_tx["transaction_type"] == "SIP"].copy()
sip_tx = sip_tx.sort_values(["investor_id", "transaction_date"])

# Keep only investors with 6+ SIP transactions
sip_counts = sip_tx.groupby("investor_id").size()
eligible = sip_counts[sip_counts >= 6].index
sip_eligible = sip_tx[sip_tx["investor_id"].isin(eligible)]

# Compute avg gap between SIP transactions per investor
def avg_gap(dates):
    dates = sorted(dates)
    gaps = [(dates[i+1] - dates[i]).days for i in range(len(dates)-1)]
    return np.mean(gaps)

sip_gaps = sip_eligible.groupby("investor_id")["transaction_date"].apply(avg_gap).reset_index()
sip_gaps.columns = ["investor_id", "avg_gap_days"]

# Flag at-risk investors (gap > 35 days)
sip_gaps["status"] = sip_gaps["avg_gap_days"].apply(
    lambda x: "At-Risk" if x > 35 else "Regular"
)

# Summary
print(sip_gaps["status"].value_counts())
print(f"\nTotal eligible investors: {len(sip_gaps)}")
print(f"At-risk investors: {len(sip_gaps[sip_gaps['status'] == 'At-Risk'])}")

# Save
sip_gaps.to_csv("../data/processed/sip_continuity.csv", index=False)
print("✅ Saved: sip_continuity.csv")
```

```python
# Load data
df_perf = pd.read_csv("../data/processed/clean_performance.csv")
df_master = pd.read_csv("../data/raw/01_fund_master.csv")

# Merge performance with risk category
df_merged = df_perf.merge(
    df_master[["amfi_code", "risk_category"]], 
    on="amfi_code"
)

def recommend_funds(risk_appetite):
    """
    Input: risk_appetite = 'Low' / 'Moderate' / 'High'
    Output: Top 3 funds by Sharpe ratio within matching risk_grade
    """
    filtered = df_merged[df_merged["risk_category"] == risk_appetite]
    top3 = filtered.nlargest(3, "sharpe_ratio")[
        ["scheme_name", "fund_house", "sharpe_ratio", 
         "return_3yr_pct", "risk_category"]
    ]
    return top3

# Test all 3 risk appetites
for risk in ["Low", "Moderate", "High"]:
    print(f"\n{'='*50}")
    print(f"Top 3 Funds for {risk} Risk Appetite:")
    print('='*50)
    print(recommend_funds(risk).to_string(index=False))
```

```python
# Load portfolio holdings
df_holdings = pd.read_csv("../data/raw/09_portfolio_holdings.csv")
print(df_holdings.head())
print(df_holdings.columns.tolist())
print(df_holdings.shape)
```

```python
# Compute Sector HHI per fund
# HHI = Sum of (sector_weight)^2 for each fund

# First aggregate weight by sector per fund
sector_weights = df_holdings.groupby(
    ["amfi_code", "sector"]
)["weight_pct"].sum().reset_index()

# Compute HHI per fund
def compute_hhi(weights):
    return (weights ** 2).sum()

hhi_df = sector_weights.groupby("amfi_code")["weight_pct"].apply(
    compute_hhi
).reset_index()
hhi_df.columns = ["amfi_code", "hhi_score"]

# Merge with fund names
hhi_df = hhi_df.merge(
    df_master[["amfi_code", "scheme_name", "category", "fund_house"]],
    on="amfi_code"
)

# Classify concentration
hhi_df["concentration"] = hhi_df["hhi_score"].apply(
    lambda x: "High" if x > 2500 else ("Moderate" if x > 1500 else "Low")
)

# Sort by HHI descending
hhi_df = hhi_df.sort_values("hhi_score", ascending=False)

print(hhi_df[["scheme_name", "hhi_score", "concentration"]].to_string(index=False))

# Save
hhi_df.to_csv("../data/processed/sector_hhi.csv", index=False)
print("\n✅ Saved: sector_hhi.csv")
```

## Advanced Analytics — 5 Key Insights

### Insight 1: Small Cap Funds Carry Highest Risk
Based on VaR analysis, Small Cap funds (ABSL, Axis, SBI Small Cap) 
show worst VaR of -2.3% to -2.4% daily, confirming highest downside 
risk. Liquid funds show near-zero VaR (-0.02%), ideal for risk-averse investors.

### Insight 2: Rolling Sharpe Shows Market Cycles Clearly
All 5 funds dipped to negative Sharpe around Jan-2024, indicating a 
market correction period. SBI Bluechip recovered strongest, peaking 
at Sharpe ~6 in mid-2025, suggesting superior risk-adjusted returns.

### Insight 3: Newer Investors (2025 Cohort) Invest More
2025 cohort investors have slightly higher avg SIP (₹1.09L) vs 2024 
cohort (₹1.07L), suggesting growing financial awareness among newer 
investors entering the market.

### Insight 4: 97.8% SIP Investors Are At-Risk of Discontinuation
Out of 1,362 eligible investors, 1,332 show avg SIP gaps > 35 days. 
This highlights a critical retention problem — AMCs should send 
reminders and incentivize consistent SIP behavior.

### Insight 5: Axis Bluechip is Most Concentrated Fund
With HHI score of 2,967, Axis Bluechip has highest sector 
concentration. UTI Mid Cap (1,240) is most diversified. Investors 
seeking diversification should prefer low HHI funds.
