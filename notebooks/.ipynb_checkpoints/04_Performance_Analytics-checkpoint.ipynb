```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# Load cleaned NAV data
df_nav = pd.read_csv("../data/processed/clean_nav.csv")
df_nav['date'] = pd.to_datetime(df_nav['date'])

# Load fund master
df_fund = pd.read_csv("../data/raw/01_fund_master.csv")

print(f"NAV data shape: {df_nav.shape}")
print(f"Fund master shape: {df_fund.shape}")
print(f"Date range: {df_nav['date'].min()} to {df_nav['date'].max()}")
print(f"Total funds: {df_nav['amfi_code'].nunique()}")
print("\n✅ Libraries and data loaded successfully!")
```

```python
# Task 1: Compute Daily Returns for all 40 funds
df_nav = df_nav.sort_values(['amfi_code', 'date'])

# Compute daily return within each fund group
df_nav['daily_return'] = df_nav.groupby('amfi_code')['nav'].pct_change()

# Drop nulls (first row of each fund has no previous day)
df_returns = df_nav.dropna(subset=['daily_return'])

# Validate - check distribution looks reasonable
print("Daily Return Statistics:")
print(df_returns['daily_return'].describe())

print(f"\nMax single day gain: {df_returns['daily_return'].max()*100:.2f}%")
print(f"Max single day loss: {df_returns['daily_return'].min()*100:.2f}%")
print(f"Average daily return: {df_returns['daily_return'].mean()*100:.4f}%")

# Save
df_returns.to_csv("../data/processed/returns_computed.csv", index=False)
print("\n✅ Saved → data/processed/returns_computed.csv")
```

```python
# Task 2: Compute CAGR for 1yr, 3yr, 5yr periods
from datetime import datetime

# Reference date = last date in data
end_date = df_nav['date'].max()
date_1yr = end_date - pd.DateOffset(years=1)
date_3yr = end_date - pd.DateOffset(years=3)
date_5yr = end_date - pd.DateOffset(years=5)

print(f"End date: {end_date.date()}")
print(f"1yr start: {date_1yr.date()}")
print(f"3yr start: {date_3yr.date()}")
print(f"5yr start: {date_5yr.date()}")

cagr_results = []

for code in df_nav['amfi_code'].unique():
    fund_data = df_nav[df_nav['amfi_code'] == code].sort_values('date')
    
    nav_end = fund_data[fund_data['date'] == fund_data['date'].max()]['nav'].values
    if len(nav_end) == 0:
        continue
    nav_end = nav_end[0]
    
    # Get fund name
    name = df_fund[df_fund['amfi_code'] == code]['scheme_name'].values
    name = name[0] if len(name) > 0 else str(code)
    
    def get_cagr(start_date, n_years):
        start_data = fund_data[fund_data['date'] >= start_date]
        if len(start_data) == 0:
            return None
        nav_start = start_data.iloc[0]['nav']
        if nav_start == 0:
            return None
        return (nav_end / nav_start) ** (1/n_years) - 1

    cagr_results.append({
        'amfi_code': code,
        'scheme_name': name,
        'cagr_1yr': get_cagr(date_1yr, 1),
        'cagr_3yr': get_cagr(date_3yr, 3),
        'cagr_5yr': get_cagr(date_5yr, 5)
    })

df_cagr = pd.DataFrame(cagr_results)

# Convert to percentage
df_cagr['cagr_1yr_pct'] = (df_cagr['cagr_1yr'] * 100).round(2)
df_cagr['cagr_3yr_pct'] = (df_cagr['cagr_3yr'] * 100).round(2)
df_cagr['cagr_5yr_pct'] = (df_cagr['cagr_5yr'] * 100).round(2)

print("\nTop 5 Funds by 3yr CAGR:")
print(df_cagr[['scheme_name', 'cagr_1yr_pct', 'cagr_3yr_pct', 'cagr_5yr_pct']]
      .sort_values('cagr_3yr_pct', ascending=False).head())

df_cagr.to_csv("../data/processed/cagr_report.csv", index=False)
print("\n✅ Saved → data/processed/cagr_report.csv")
```

```python
# Task 3: Compute Sharpe Ratio for all 40 funds
Rf = 0.065 / 252  # Daily risk free rate (6.5% annual / 252 trading days)

sharpe_results = []

for code in df_returns['amfi_code'].unique():
    fund_returns = df_returns[df_returns['amfi_code'] == code]['daily_return']
    
    # Get fund name
    name = df_fund[df_fund['amfi_code'] == code]['scheme_name'].values
    name = name[0] if len(name) > 0 else str(code)
    
    # Sharpe Ratio
    excess_return = fund_returns.mean() - Rf
    std_return = fund_returns.std()
    
    if std_return == 0:
        sharpe = None
    else:
        sharpe = (excess_return / std_return) * np.sqrt(252)
    
    sharpe_results.append({
        'amfi_code': code,
        'scheme_name': name,
        'avg_daily_return': fund_returns.mean(),
        'std_daily_return': std_return,
        'sharpe_ratio': sharpe
    })

df_sharpe = pd.DataFrame(sharpe_results)
df_sharpe['sharpe_ratio'] = df_sharpe['sharpe_ratio'].round(3)

print("Top 5 Funds by Sharpe Ratio:")
print(df_sharpe[['scheme_name', 'sharpe_ratio']]
      .sort_values('sharpe_ratio', ascending=False).head())

print(f"\nFunds with Sharpe > 1: {len(df_sharpe[df_sharpe['sharpe_ratio'] > 1])}")
print(f"Funds with Sharpe < 0: {len(df_sharpe[df_sharpe['sharpe_ratio'] < 0])}")

df_sharpe.to_csv("../data/processed/sharpe_values.csv", index=False)
print("\n✅ Saved → data/processed/sharpe_values.csv")
```

```python
# Task 4: Compute Sortino Ratio
sortino_results = []

for code in df_returns['amfi_code'].unique():
    fund_returns = df_returns[df_returns['amfi_code'] == code]['daily_return']
    
    # Get fund name
    name = df_fund[df_fund['amfi_code'] == code]['scheme_name'].values
    name = name[0] if len(name) > 0 else str(code)
    
    # Sortino Ratio
    excess_return = fund_returns.mean() - Rf
    
    # Only negative returns for downside std
    negative_returns = fund_returns[fund_returns < 0]
    downside_std = negative_returns.std()
    
    if downside_std == 0 or pd.isna(downside_std):
        sortino = None
    else:
        sortino = (excess_return / downside_std) * np.sqrt(252)
    
    sortino_results.append({
        'amfi_code': code,
        'scheme_name': name,
        'sortino_ratio': sortino
    })

df_sortino = pd.DataFrame(sortino_results)
df_sortino['sortino_ratio'] = df_sortino['sortino_ratio'].round(3)

print("Top 5 Funds by Sortino Ratio:")
print(df_sortino[['scheme_name', 'sortino_ratio']]
      .sort_values('sortino_ratio', ascending=False).head())

df_sortino.to_csv("../data/processed/sortino_values.csv", index=False)
print("\n✅ Saved → data/processed/sortino_values.csv")
```

```python
# Task 5: Compute Alpha & Beta vs Nifty 100
df_benchmark = pd.read_csv("../data/raw/10_benchmark_indices.csv")
df_benchmark['date'] = pd.to_datetime(df_benchmark['date'])
print("Benchmark columns:", df_benchmark.columns.tolist())
print(df_benchmark.head())
```

```python
# Task 5: Compute Alpha & Beta vs Nifty 100

# Get Nifty 100 data
nifty100 = df_benchmark[df_benchmark['index_name'] == 'NIFTY100'].copy()

# If NIFTY100 not available, use NIFTY50
if len(nifty100) == 0:
    print("NIFTY100 not found, using NIFTY50 instead")
    nifty100 = df_benchmark[df_benchmark['index_name'] == 'NIFTY50'].copy()

# Compute benchmark daily returns
nifty100 = nifty100.sort_values('date')
nifty100['benchmark_return'] = nifty100['close_value'].pct_change()
nifty100 = nifty100.dropna()

print(f"Benchmark data: {len(nifty100)} rows")
print(f"Benchmark indices available: {df_benchmark['index_name'].unique()}")

alpha_beta_results = []

for code in df_returns['amfi_code'].unique():
    fund_returns = df_returns[df_returns['amfi_code'] == code][['date', 'daily_return']]
    fund_returns['date'] = pd.to_datetime(fund_returns['date'])
    
    # Get fund name
    name = df_fund[df_fund['amfi_code'] == code]['scheme_name'].values
    name = name[0] if len(name) > 0 else str(code)
    
    # Merge fund returns with benchmark returns on date
    merged = pd.merge(fund_returns, nifty100[['date', 'benchmark_return']], 
                      on='date', how='inner')
    
    if len(merged) < 30:
        continue
    
    # OLS Regression
    slope, intercept, r_value, p_value, std_err = stats.linregress(
        merged['benchmark_return'], 
        merged['daily_return']
    )
    
    beta = slope
    alpha = intercept * 252  # Annualise alpha
    
    alpha_beta_results.append({
        'amfi_code': code,
        'scheme_name': name,
        'alpha': round(alpha, 4),
        'beta': round(beta, 4),
        'r_squared': round(r_value**2, 4)
    })

df_alpha_beta = pd.DataFrame(alpha_beta_results)

print("\nTop 5 Funds by Alpha:")
print(df_alpha_beta[['scheme_name', 'alpha', 'beta']]
      .sort_values('alpha', ascending=False).head())

df_alpha_beta.to_csv("../data/processed/alpha_beta.csv", index=False)
print("\n✅ Saved → data/processed/alpha_beta.csv")
```

```python
# Task 6: Compute Maximum Drawdown
drawdown_results = []

for code in df_nav['amfi_code'].unique():
    fund_data = df_nav[df_nav['amfi_code'] == code].sort_values('date')
    
    # Get fund name
    name = df_fund[df_fund['amfi_code'] == code]['scheme_name'].values
    name = name[0] if len(name) > 0 else str(code)
    
    # Calculate running maximum
    fund_data['running_max'] = fund_data['nav'].cummax()
    
    # Calculate drawdown
    fund_data['drawdown'] = (fund_data['nav'] / fund_data['running_max']) - 1
    
    # Maximum drawdown
    max_dd = fund_data['drawdown'].min()
    
    # Find worst drawdown date
    worst_date = fund_data.loc[fund_data['drawdown'].idxmin(), 'date']
    
    drawdown_results.append({
        'amfi_code': code,
        'scheme_name': name,
        'max_drawdown_pct': round(max_dd * 100, 2),
        'worst_date': worst_date
    })

df_drawdown = pd.DataFrame(drawdown_results)

print("Top 5 Funds with Worst Drawdown:")
print(df_drawdown[['scheme_name', 'max_drawdown_pct', 'worst_date']]
      .sort_values('max_drawdown_pct').head())

print("\nTop 5 Funds with Best (Least) Drawdown:")
print(df_drawdown[['scheme_name', 'max_drawdown_pct', 'worst_date']]
      .sort_values('max_drawdown_pct', ascending=False).head())

df_drawdown.to_csv("../data/processed/max_drawdown.csv", index=False)
print("\n✅ Saved → data/processed/max_drawdown.csv")
```

```python
# Task 7: Build Fund Scorecard
df_perf = pd.read_csv("../data/raw/07_scheme_performance.csv")

# Merge all metrics
df_score = df_perf[['amfi_code', 'scheme_name', 'fund_house',
                     'return_3yr_pct', 'sharpe_ratio', 
                     'expense_ratio_pct', 'max_drawdown_pct']].copy()

# Merge alpha
df_score = df_score.merge(
    df_alpha_beta[['amfi_code', 'alpha']], 
    on='amfi_code', how='left'
)

# Rank each metric (higher rank = better)
df_score['rank_3yr_return'] = df_score['return_3yr_pct'].rank(pct=True)
df_score['rank_sharpe'] = df_score['sharpe_ratio'].rank(pct=True)
df_score['rank_alpha'] = df_score['alpha'].rank(pct=True)

# Inverse ranks (lower expense ratio = better, lower drawdown = better)
df_score['rank_expense'] = (1 - df_score['expense_ratio_pct'].rank(pct=True))
df_score['rank_maxdd'] = (1 - df_score['max_drawdown_pct'].rank(pct=True))

# Composite score
df_score['composite_score'] = (
    0.30 * df_score['rank_3yr_return'] +
    0.25 * df_score['rank_sharpe'] +
    0.20 * df_score['rank_alpha'] +
    0.15 * df_score['rank_expense'] +
    0.10 * df_score['rank_maxdd']
) * 100

df_score['composite_score'] = df_score['composite_score'].round(2)

# Final scorecard
df_scorecard = df_score[['amfi_code', 'scheme_name', 'fund_house',
                          'return_3yr_pct', 'sharpe_ratio', 'alpha',
                          'expense_ratio_pct', 'max_drawdown_pct',
                          'composite_score']].sort_values(
                          'composite_score', ascending=False)

print("🏆 TOP 10 FUNDS BY SCORECARD:")
print("="*60)
print(df_scorecard[['scheme_name', 'composite_score']].head(10).to_string())

df_scorecard.to_csv("../data/processed/fund_scorecard.csv", index=False)
print("\n✅ Saved → data/processed/fund_scorecard.csv")
```

```python
# Task 8: Benchmark Comparison Chart
# Get top 5 funds from scorecard
top5_codes = df_scorecard.head(5)['amfi_code'].tolist()

# Get NAV data for top 5 funds
df_top5 = df_nav[df_nav['amfi_code'].isin(top5_codes)].copy()

# Merge with fund names
df_top5 = df_top5.merge(
    df_fund[['amfi_code', 'scheme_name']], 
    on='amfi_code', how='left'
)

# Get Nifty 50 and Nifty 100 benchmark data
nifty50 = df_benchmark[df_benchmark['index_name'] == 'NIFTY50'].copy()
nifty100 = df_benchmark[df_benchmark['index_name'] == 'NIFTY100'].copy()

# Normalise everything to 100 at start date
start_date = '2023-01-01'

def normalise(df, date_col, value_col, name):
    df = df[df[date_col] >= start_date].copy()
    base = df.iloc[0][value_col]
    df['normalised'] = (df[value_col] / base) * 100
    df['name'] = name
    return df[[date_col, 'normalised', 'name']]

# Plot
plt.figure(figsize=(10, 6))

# Plot top 5 funds
for code in top5_codes:
    fund_data = df_top5[df_top5['amfi_code'] == code]
    name = fund_data['scheme_name'].iloc[0][:25]
    fund_norm = fund_data[fund_data['date'] >= start_date].copy()
    if len(fund_norm) == 0:
        continue
    base = fund_norm.iloc[0]['nav']
    fund_norm['normalised'] = (fund_norm['nav'] / base) * 100
    plt.plot(fund_norm['date'], fund_norm['normalised'], 
             linewidth=1.5, label=name)

# Plot benchmarks
nifty50_norm = nifty50[nifty50['date'] >= start_date].copy()
base50 = nifty50_norm.iloc[0]['close_value']
nifty50_norm['normalised'] = (nifty50_norm['close_value'] / base50) * 100
plt.plot(nifty50_norm['date'], nifty50_norm['normalised'],
         linewidth=2.5, linestyle='--', color='black', label='Nifty 50')

nifty100_norm = nifty100[nifty100['date'] >= start_date].copy()
base100 = nifty100_norm.iloc[0]['close_value']
nifty100_norm['normalised'] = (nifty100_norm['close_value'] / base100) * 100
plt.plot(nifty100_norm['date'], nifty100_norm['normalised'],
         linewidth=2.5, linestyle=':', color='red', label='Nifty 100')

plt.title('Top 5 Funds vs Benchmark (Normalised to 100)\nJan 2023 - May 2026',
          fontsize=14, fontweight='bold')
plt.xlabel('Date', fontsize=12)
plt.ylabel('Normalised Value (Base=100)', fontsize=12)
plt.legend(fontsize=8)
plt.grid(True, alpha=0.3)
plt.tight_layout()

plt.savefig('../reports/charts/10_benchmark_comparison.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ Chart saved → reports/charts/10_benchmark_comparison.png")
```
